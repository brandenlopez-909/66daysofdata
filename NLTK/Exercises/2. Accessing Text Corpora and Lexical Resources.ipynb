# 14 Define a function supergloss(s) that takes a synset s as its argument and returns a string consisting of the concatenation of the definition of s, and the definitions of all the hypernyms and hyponyms of s.

```python
def meronymsHolonyms(word):
    from nltk.corpus import wordnet as wn
    print("The member meronyms for " + str(word) + " are " + str(wn.synset(word).member_meronyms() )  )
    print( "The part_meronyms for " + str(word) + " are " + str(wn.synset(word).part_meronyms() ) )
    print( "The substance_meronyms for " + str(word) + " are " + str(wn.synset(word).substance_meronyms()) )
    print( "The member_holonyms for " + str(word) + " are " +  str(wn.synset(word).member_holonyms()) )
    print("The part_holonyms for " +str( word) + " are " +  str(wn.synset(word).part_holonyms()) )
    print("The substance_holonyms for " + str(word) + " are " +  str(wn.synset(word).substance_holonyms()) )

#Supergloss requires the input to be in syssetform
def supergloss(s):
    from nltk.corpus import wordnet as wn
    S =wn.synset(s)
    typeOFS = S.hyponyms()
    print("The hyponyms are : \n")
    for word in typeOFS:
        print("\t" + str(word) + word.definition() )
        
    typeOFS = S.hypernyms()
    print("\nThe hypernyms are : \n")
    for word in typeOFS:
        print("\t" + str(word) + word.definition() )
    
    
```

# 20. Write a function word_freq() that takes a word and the name of a section of the Brown Corpus as arguments, and computes the frequency of the word in that section of the corpus.

```python
def word_freq(word, category):
    word_types = [w.lower() for w in nltk.corpus.brown.words(categories=category) if w.isalpha()]
    occurances = [w.lower() for w in nltk.corpus.brown.words(categories=category) if w.lower() == word.lower()]
    return len(occurances)/ len(word_types)
```

```python
def hedge(text):
    n = 3
    like = list(text.split())

    while n <= len(like):
        like.insert(n, 'like')
        n = n + 4

    return like
```
